```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
def data_split(vectors_df, X_label='X_tfidf', random_state=12):
    X = vectors_df[X_label].iat[0]
    X_train, X_test, y_train, y_test = train_test_split(X, reviews_df['ratings'], test_size=0.33, random_state=random_state)
    ## x is scipy.sparse.csr.csr_matrix and y is  pandas.core.series.Series,
    # [type(y) for y in [X_train, X_test, y_train, y_test]]
    return X_train, X_test, y_train, y_test

def add_data_splits(vectors_df):
    test_train_splits_df = pd.DataFrame()
    random_state=12
    labels, X_trains, X_tests, y_trains, y_tests = list(), list(), list(), list(), list()
    for label in vectors_df.columns:
        X_train, X_test, y_train, y_test = data_split(vectors_df, label, random_state)
        labels.append(label)
        X_trains.append(X_train)
        X_tests.append(X_test)
        y_trains.append(y_train)
        y_tests.append(y_test)
    test_train_splits_df['label'] = labels
    test_train_splits_df['random_state'] = random_state
    test_train_splits_df['X_train'] = X_trains
    test_train_splits_df['X_test'] = X_tests
    test_train_splits_df['y_train'] = y_trains
    test_train_splits_df['y_test'] = y_tests
    return test_train_splits_df
```

```python
def main():
    reviews_df = pd.read_pickle('./data/reviews_df.pickle')
    vectors_df = pd.read_pickle('./data/vectors_df.pickle')
    test_train_splits_df = add_data_splits(vectors_df)
    test_train_splits_df.to_pickle('./data/test_train_splits_df.pickle')
    test_train_splits_df_loaded= pd.read_pickle('./data/test_train_splits_df.pickle')
    return test_train_splits_df_loaded.head()
```

```python
main()
```
